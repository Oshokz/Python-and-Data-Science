### Data Cleaning 2 - Dealing with missing values & tidying of data

#### Missing Data
- In databases, they are NULL values - there is nothing there i.e. Blank cell. it is rep by NAN. Empty string is not NAN
- There is no one best approach to handle missing values. It may require adequate knowledge of the data at hand. 
However, sometimes we might need to fill some NAs with a zero, average or median of the column, or drop the 
column with the missing values. Note that the latter is not ideal as we might be losing great information.


Methods
1. Deletion (pairwise, list wise(entire rows) and dropping of columns): Deleting missing value should be the last option; because for every data you delete, you are losing information of people of transactions except that very few rows or almost all rows are missing and deleting it wont have much effect. all the the types are not recommended.
2. Imputation: this is the process of filling the missing values. Two ways <br>
a. general - for time series and non time series: <br>
i.time series: a data indexed with date/time collected at a specific interval e.g. data will be collected every 1hr <br>
or every 30 sec; imputation techniques include forward fill, backward fill, linear intorpolation <br>
stock market data, weather forecast, transaction with data. Linear interpolation: if the price of stock market is this <br>
and in 30 seconds, the price should not have changed too much so we can infer it <br>
ii.non-time series:most common. it is not indexed with time. not collected at any interval the data can be collected <br>
for the past 2 years and there is nothing like data coming every minute or every hour but it might have dates in it <br>
imputation techniques include imputing with a constant, mean, median or mode <br>
b. advanced: for time series;using sci-kit learn - KNIN based and MICE <br>

```python
# we define a function to convert the duration column to numeric  [1 * 60 + 47]
def time_to_minute(var): # var is representing each row of the dataset
    if isinstance(var, str):# excluding the missing values i.e. NAN
        # check cases of hr only first
        if "hr" in var and "min" not in var: #the second condition is for it to exclude the duration that has both hr and min
            hr= float(var.replace("hr",""))*60
            return hr
        # check cases of min only 
        elif "min" in var and "hr" not in var:
            mins= float(var.replace("min",""))
            return mins
        # hr and min together
        elif "hr" in var and "min" in var:
            hr = float(var.split()[0].replace("hr","")) * 60
            mins = float(var.split()[1].replace("min",""))
            return hr + mins    
    # check for NAN
    else:
        return np.nan
```

```python
# let's apply the function to the duration column and overwrite the column
df["duration"] = df["duration"].apply(time_to_minute)
```

### Utility function 1

```python
# Defining the function for creating boxplot and hisogram 
def histogram_boxplot(data, feature, figsize=(9, 5), kde=False, bins=None):
    """
    Boxplot and histogram combined

    data: dataframe
    feature: dataframe column
    figsize: size of figure (default (12,7))
    kde: whether to show the density curve (default False)
    bins: number of bins for histogram (default None)
    """
    f2, (ax_box2, ax_hist2) = plt.subplots(
        nrows=2,  # Number of rows of the subplot grid= 2
        sharex=True,  # x-axis will be shared among all subplots
        gridspec_kw={"height_ratios": (0.25, 0.75)},
        figsize=figsize)  # creating the 2 subplots
    
    sns.boxplot(data=data, x=feature, ax=ax_box2, showmeans=True, color="mediumturquoise")  # boxplot will be created and a star will indicate the mean value of the column
    
    if bins:
      sns.histplot(data=data, x=feature, kde=kde, ax=ax_hist2, bins=bins, color="mediumpurple")
    else: 
      sns.histplot(data=data, x=feature, kde=kde, ax=ax_hist2, color="mediumpurple")  # For histogram
    
    ax_hist2.axvline(data[feature].mean(), color="green", linestyle="--")  # Add mean to the histogram
    
    ax_hist2.axvline(data[feature].median(), color="black", linestyle="-")  # Add median to the histogram
```

### Utility Function 2 

```python
# function to create labeled barplots

def labeled_barplot(data, feature, perc=False, n=None):
    """
    Barplot with percentage at the top

    data: dataframe
    feature: dataframe column
    perc: whether to display percentages instead of count (default is False)
    n: displays the top n category levels (default is None, i.e., display all levels)
    """

    total = len(data[feature])  # length of the column
    count = data[feature].nunique()
    if n is None:
        plt.figure(figsize=(count + 1, 5))
    else:
        plt.figure(figsize=(n + 1, 5))

    plt.xticks(rotation=90, fontsize=15) # this is the plot
    ax = sns.countplot(
        data=data,
        x=feature,
        palette="Paired", #how it will display the count plot
        order=data[feature].value_counts().index[:n].sort_values(), #sorting
    )

    for p in ax.patches: # to calculate the percentage
        if perc == True:
            label = "{:.1f}%".format(
                100 * p.get_height() / total
            )  # percentage of each class of the category
        else:
            label = p.get_height()  # count of each level of the category

        x = p.get_x() + p.get_width() / 2  # width of the plot
        y = p.get_height()  # height of the plot

        ax.annotate( #to put percentage on top of the bar
            label,
            (x, y),
            ha="center",
            va="center",
            size=12,
            xytext=(0, 5),
            textcoords="offset points",
        )  # annotate the percentage

    plt.show()  # show the plot
```

#### Splitting the data into the training and test set
the essence of buildinga model is to use it to make prediction on new and unseen data; hence there is a need to split the data into training and validation/testset. So we can buind the model with training set and validate it with test set <br>

Scikit learn is the standard library to build machine learning model; there is a function in it that can help us to split the data. it randomly chooses rows for data in test and train

```python
from sklearn.model_selection import train_test_split

# split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
#"""random_state is a seeed that wiill prevent the machine from loading different set of rows in case the code is leaded again. 
#SO, everytime you run this cell, it is going to give you the exact data that has been run before""" 
```

#### Fitting the linear regression model

```python
from sklearn.linear_model import LinearRegression
# at this point LinearRegression is an algorithm because it has not yet been trained on a data
# after it has been trained on a data, it is no longer an algorithm now a model because it has come up with a function
# that can be used to make prediction

# instatiate the linear regression algorithm
lr = LinearRegression()

# fit the model: this means train the model i.e. supply data to the algorithm and then it will come up with a model
lr.fit(X_train, y_train)

# get the weight or coefficient and the intercept or bias or offset level
print(lr.coef_)
```

 the model has 0.44591367 coefficient. the more appropriate the coefficient the more the model will perform 

```python
# training and test accuracy
print(lr.score(X_train, y_train))
```

```python
print("Train Accuracy: {}" .format(lr.score(X_train, y_train)))
```

```python
print("Test Accuracy: {:.4f}" .format(lr.score(X_test, y_test)))
```

Overfitting (high variance): means the model performed very well on the data used to build it( train data) but worst on the test data; remember our focus is more on the test. It means the model overfit the training data. it means it memeorizes all the patterns including the noise (things not needed in the data) in the data and it was too weak to perform very well when we gave it a new data it hasnt seen before. Such model will give very high accuracy in the trainingso if you dont check it on the test, you will think it is doing well. That is why it is important to always check the accuracy on the testing data too

underfitting (high bias): when the model is too weak to perform well on both the train and test. it is too weak to understand the patterns in the data. the model is too simple to understand the pattern in the training and as a result, it gives low accuracy in both training and testing

good compromise: model performed well on the training and much more on the testing (higher): so it generalizes well. for instance our model is a good compromise. This is because it could get to this level even with one feature. because you are building the model for future, you have to focus on test data. Test data is a data the model hasnt seen before. You put more emphasis on the test data because it is an example of the future data that will be similar to the data used to build it and if the test data accuracy is higher then the model generalizes well
